```python
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_numerical = X_Numerical.reset_index()
```

```python
New_Bike = pd.concat([X_cat, X_numerical], axis = 1)
```

```python
New_Bike = New_Bike.drop('dteday', axis = 1)
```

```python
New_Bike
```

```python
X = New_Bike.iloc[:, :-1].values
y = New_Bike.iloc[:, -1:].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y = scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=100, activation='relu', input_shape=(35, )))
model.add(tf.keras.layers.Dense(units=200, activation='relu'))
model.add(tf.keras.layers.Dense(units=200, activation='relu'))
model.add(tf.keras.layers.Dense(units=100, activation='relu'))
model.add(tf.keras.layers.Dense(units=200, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 200, batch_size = 50, validation_split = 0.2)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation Loss')
plt.legend(['Training Loss', 'Validation Loss'])
```

```python
y_predict = model.predict(X_test)
plt.plot(y_test, y_predict, "x", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```
